```python
#function plot wordcloud
def plotwcloud(wordcloud):
    # size of figure
    plt.figure(figsize=(30, 50))
    # show pic
    plt.imshow(wordcloud) 
    # without labels
    plt.axis("off")
```

```python
#create string
text_list=''
for row in data_1[data_1['Name']=='–¢–∞—Ç—å—è–Ω–∞']['Message']:
    text_list=text_list+' '+row
```

```python
from wordcloud import WordCloud
from stop_words import get_stop_words

#variable for stopword
stopw = get_stop_words('russian')

# create wordcloud
mycloud = WordCloud(width = 2000, 
                      height = 1500, 
                      random_state=1, 
                      background_color='black', 
                      margin=20, 
                      colormap='Pastel1', 
                      collocations=False, 
                      stopwords = stopw).generate(text_list)

# drow pic
plotwcloud(mycloud)
mycloud.to_file('hp_cloud_simple.png')
```
